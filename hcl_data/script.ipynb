# Source 1: HCL DATA

HCL RSS processing pipeline and HCL lab processing pipeline are in the same Kedro pipeline (and same code repository)

#### Step 1: Build the Docker container
```bash
kedro docker build
```
#### Step 2: Set environment variables (replace placeholders with your credentials)
```bash
export AWS_ACCESS_KEY_ID="YOUR_AWS_ACCESS_KEY_ID"
export AWS_SECRET_ACCESS_KEY="YOUR_AWS_SECRET_ACCESS_KEY"
export MongoDBUsername="YOUR_MongoDBUsername"
export MongoDBPassword="YOUR_MongoDBPassword"
```
#### Step 3: Run pipelines with Docker container in the 'prod' environment

pipeline_name in the command line below must be replaced by: 
- data_preprocessing_rss : for HCL RSS data
- data_preprocessing_lab : for HCL Lab data

```bash
kedro docker run --pipeline=pipeline_name --env=prod --docker-args="--env AWS_ACCESS_KEY_ID=AWS_ACCESS_KEY_ID --env AWS_SECRET_ACCESS_KEY=AWS_SECRET_ACCESS_KEY --env MongoDBUsername=MongoDBUsername --env MongoDBPassword=MongoDBPassword --env CGM_PASSWORD=CGM_PASSWORD"
```
#### Step 4: Run all pipelines with Docker container in the 'prod' environment (HCL RSS + HCL Lab)
```bash
kedro docker run --env=prod --docker-args="--env AWS_ACCESS_KEY_ID=AWS_ACCESS_KEY_ID --env AWS_SECRET_ACCESS_KEY=AWS_SECRET_ACCESS_KEY --env MongoDBUsername=MongoDBUsername --env MongoDBPassword=MongoDBPassword --env CGM_PASSWORD=CGM_PASSWORD"
```

# Source 3: Covid RKI DATA

#### Step 1: Build the Docker container
```bash
kedro docker build
```
#### Step 2: Set environment variables (replace placeholders with your credentials)
```bash
export AWS_ACCESS_KEY_ID="YOUR_AWS_ACCESS_KEY_ID"
export AWS_SECRET_ACCESS_KEY="YOUR_AWS_SECRET_ACCESS_KEY"
export MongoDBUsername="YOUR_MongoDBUsername"
export MongoDBPassword="YOUR_MongoDBPassword"
```
#### Step 3: Run the RKI DATA pipeline with Docker container in 'prod' environment
```bash
kedro docker run --pipeline=web_data_scrapping --env=prod --docker-args="--env AWS_ACCESS_KEY_ID=$AWS_ACCESS_KEY_ID --env AWS_SECRET_ACCESS_KEY=$AWS_SECRET_ACCESS_KEY --env MongoDBUsername=$MongoDBUsername --env MongoDBPassword=$MongoDBPassword"
```
#### Step 4: Optionally, run the pipeline locally for prodelopment (outside Docker)
```bash
kedro run --env=prod
```
#### Additional Commands for RKI DATA

#### To get the latest RKI data
```bash
kedro docker run --pipeline=web_data_scrapping --env=prod
```
##### For raw data validation
```bash
kedro docker run --pipeline=extracted_data_validation --env=prod
```
##### For ABM calibration
```bash
kedro docker run --pipeline=model_calibration --env=prod
```
##### Generate predictions using the calibrated model
```bash
kedro docker run --pipeline=model_predictions --env=prod
```
